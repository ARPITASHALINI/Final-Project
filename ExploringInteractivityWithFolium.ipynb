```python
from branca.element import Figure                                               # resizing the map by using Branca library in python. 
fig = Figure(width = 600, height = 450)
```

```python
#Setting minimum and maximum zoom levels
map1 = folium.Map(width=600, height=450, location=[42.27199589252885, -71.80396347794411], zoom_start=11, min_zoom=8, max_zoom=14)             
fig.add_child(map1)
map1 
```

```python
fig2 = Figure(width = 600, height = 450)                                        # In this chunk we give an option to chose between different basemaps
map2 = folium.Map(location = [42.27199589252885, -71.80396347794411])           # Setting the zoom to Worcester. 
fig2.add_child(map2)
folium.TileLayer('Stamen Terrain').add_to(map2)                                 # Stamen Terrain as an option for basemap
folium.TileLayer('Stamen Toner').add_to(map2)                                   # Stamen Toner as an option for basemap
folium.TileLayer('Stamen Water Color').add_to(map2)                             # Stamen Water Color as an option for basemap  
folium.TileLayer('cartodbpositron').add_to(map2)                                # CartoDB Positron as an option for basemap
folium.TileLayer('cartodbdark_matter').add_to(map2)                             # CartoDB Dark Matter as an option for basemap
folium.LayerControl().add_to(map2)                                              # Bar to chose between different layers. 
map2
```
